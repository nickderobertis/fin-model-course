# A First `xlwings` Project

## Overview

An `xlwings` project consists of at least two files: the Excel workbook and a Python file. These two files must have the same file name, besides the extension. The extension for the Excel file will be `.xlsm` as it is a macro-enabled workbook. The extension for the Python file will be `.py`, as it is a plain Python file and not a Jupyter notebook.

Note: I say at least two, because you could have other Python files, you can `import` them into the main Python file.

## Creating the Project

Let's create a project called `first_project`. `xlwings` provides a tool to create the project using the command line. First we want to get a terminal in the folder where we want the project. To do this, you can navigate to the folder you want in Windows explorer (click through the folders), then inside the folder, hold `Shift` and right click. You should see the following:

Note: It is also possible to just create the files manually, but there is some extra setup involved which is outside the scope of this course.

You want to click "Open command window here". On Windows 10, it may say "Open Powershell window here", but it is the same thing, your terminal will just be blue instead of black.

Then you will be in a terminal. From there, you want to run `xlwings quickstart first_project`. In the future, you'll want to replace `first_project` with whatever you want to call the project. After running it, you should see something like this:

Now you should see a folder `first_project` created wherever you wanted the project. Inside it, there should be a `.xlsm` and `.py` file with the same name `first_project`. Double-click the `first_project.xlsm` to open it, and in Jupyter, navigate to that folder and open the `.py` file there.
